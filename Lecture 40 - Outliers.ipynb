```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
# Let's see how we would find outliers in a dataset

# First we'll seed the numpy generator
np.random.seed(12345)
```

```python
#Next we'll create the dataframe

#input: 1000 rows, 4 colummns
dframe = DataFrame(np.random.randn(1000,4))
```

```python
#see first few rows
dframe.head()
```

```python
#see last few rows
dframe.tail()
```

```python
# Lets describe the data
dframe.describe()
```

```python
#take first column (index-0 column)
col = dframe[0]
```

```python
#see first few values of that column
col.head()
```

```python
# Now we can check which values in the column have absolute values that are greater than 3, for instance.
col[np.abs(col)>3]
```

```python
# So we now know in column[0], rows 523 and 900 have values with abs > 3

#How about all the columns?

# We can use the "any" method
dframe[(np.abs(dframe)>3).any(1)]
#any(axis) => 1-axis (column) => if see values with abs value greater than 3 in any column, return the rows
```

```python
# WE could also possibly cap the data at 3
# one good way to cap outliers"
dframe[np.abs(dframe)>3]=np.sign(dframe)*3
#any value with abs value >3, replace that value with 1*3 or -1*3 depending on the sign of original value
```

```python
dframe.describe()
```
