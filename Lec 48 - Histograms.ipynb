```python
# Let's begin by importing what we'll need (You'll probably be copying and pasting this a lot)

# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats libray from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
#First of all, source of information for what a histogram actually is: 
# http://en.wikipedia.org/wiki/Histogram
```

```python
#Create a random normal-distribution dataset with 100 data points
dataset1 = randn(100)
```

```python
#Plot a histogram of the dataset1, note bins=10 by default

#plt is matplotlib.pyplot
plt.hist(dataset1)
```

```python
# Lets make another dataset with 80 data points
dataset2 = randn(80)

#plot the dataset2, choose color = indian red
plt.hist(dataset2,color = 'indianred')
```

```python
# We can use parameter "normed" (normalized) to plot on same plot

#dataset1 has 100 data points, dataset2 has 80 points, so you have to normalize the plots

# Set normed=True for the plots to be normalized in order to comapre data sets with different number of observations

# Set argument alpha=0.5 for transperancy

plt.hist(dataset1, normed=True, color = 'indianred', alpha = 0.5, bins = 20)
plt.hist(dataset2, normed=True, alpha = 0.5, bins = 20)

# 2 plots are stacked on top each other
```

```python
# Make two more random normal dist data sets

#dataset 1 has 1000 points
data1 = randn(1000)

#dataset 2 has 1000 points
data2 = randn(1000)
```

```python
#Can represent joint distributions using "joint plot"
# method .jointplot()
# library seaborn: sns
sns.jointplot(data1,data2)
```

```python
#notice p value: p = 0.18

# Can also use hex bins for a more concise picture
sns.jointplot(data1,data2,kind='hex')
```
