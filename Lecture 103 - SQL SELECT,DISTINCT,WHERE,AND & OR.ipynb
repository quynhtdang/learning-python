```python
# two Imports
import sqlite3
import pandas as pd

#connect to the database
con = sqlite3.connect("sakila.db")
```

```python
#make a function that automatically takes in our query and returns a DataFrame.
# Set function as our sql_to_pandasdataframe

def sql_to_df(sql_query):
    # Use pandas to pass sql query using connection form SQLite3
    df = pd.read_sql(sql_query, con)
    
    #show the resulting DataFrame
    return df
```

```python
# Select multiple columns: first_name
query = ''' SELECT first_name, last_name FROM customer;'''
```

```python
#convert those two columns in the 'customer' table to a "virtual" dataframe
#in other words, grab from first two columns of the 'customer' table
sql_to_df(query)
```

```python
#you can also preview the first 5 columns only 
sql_to_df(query).head()
```

```python
#In a table, a column may contain duplicate values; 
#and sometimes you only want to list the distinct (unique) values. 
#DISTINCT keyword in SQL is similar to .unique() method in pandas dataframe
#syntax to return only distinct (unique) values

#SELECT DISTINCT column_name 
#FROM table_name;
```

```python
#example: select distinct values from (country_id) column from the 'city' table.

query = ''' SELECT DISTINCT country_id
            FROM city'''

sql_to_df(query).head()
```

```python
#Syntax for the SQL WHERE

#The WHERE clause is used to filter records, 
#the WHERE clause is used to extract only the records that fulfill the specific "parameter".

#SELECT column_name 
#FROM table_name 
#WHERE column_name ( math operator) desired_value;
```

```python
# Select all customer info from the 1st store (column 'store_id' =  1) from the 'customer' table.
query3 = ''' SELECT *
             FROM customer
             WHERE store_id = 1'''

df1 = sql_to_df(query3)

#the function automatically takes in our query and returns a DataFrame.
#we can assign the result to an "actual" dataframe called 'df1'
```

<B>Syntax for AND</B>

The AND operator is used to filter records based on more than one condition.

The AND operator displays a record if both the first condition AND the second condition are true.

```python
# Select all films from 2006 that are rated R.
#condition 1: column 'release_year' = 2006
#condition 2: column 'rating' = 'R'
#AND => both conditions must be true
query = ''' SELECT *
            FROM film
            WHERE release_year = '2006'
            AND rating = 'R'  '''

sql_to_df(query).head()
```

<b>Syntax for OR</b>

The OR operator displays a record if either the first condition OR the second condition is true.

```python
# Select all films from R "or" PG.
# condition 1: column 'rating' = 'PG'
#OR
# condition 2: column 'rating' = 'R'

query = ''' SELECT *
            FROM film
            WHERE rating = 'PG'
            OR rating = 'R'  '''

sql_to_df(query).head()
```
